# **IT hardware and Electronics Trend Comparison Yearwise**

```python
dfIT['Date']=pd.to_datetime(dfIT['Date'])
dfelect['Date']=pd.to_datetime(dfelect['Date'])
dfIT['Sales'] = pd.to_numeric(dfIT['Sales'],errors = 'coerce')
dfITsum=dfIT.groupby('Date').sum().sort_values(['Date'],ascending=True)
dfITsum=pd.DataFrame(dfITsum)
dfelectsum=dfelect.groupby('Date').sum().sort_values(['Date'],ascending=True)
dfelectsum=pd.DataFrame(dfelectsum)
```

```python
animatedBar = pd.DataFrame(dfITsum[['Sales','Expenses','Net profit']])

bcr.bar_chart_race(animatedBar,title=' IT HARDWARE Sale Expenses and Net profit Year Wise in Crs',bar_size=.5,figsize=(7,3.5),steps_per_period=1,period_length=1500,title_size='larger',period_fmt='%b  %Y')
```

```python
animatedBar1 = pd.DataFrame(dfelectsum[['Sales','Expenses','Net profit']])

bcr.bar_chart_race(animatedBar1,title=' ELECTRONICS Sale Expenses and Net profit Year Wise in Crs',bar_size=.5,figsize=(7,3.5),steps_per_period=1,period_length=1500,title_size='larger',period_fmt='%b  %Y')
```
